```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#displaying the first 5 rows of the dataset
df = pd.read_csv(r"C:\Users\KIIT0001\Downloads\world_population.csv")
print(df.head())
```

```python
#displaying dataframe information
df.info()
```

```python
#displaying all the statistical data of the datasetdf.describe()
df.describe()
```

```python
#setting a custom float format for data display
pd.set_option('display.float_format', lambda x: '%.2f'% x)
df
```

```python
#checking for missing values in the dataset
df.isnull().sum()
```

```python
#finding the number of unique values in each column
df.nunique()
```

```python
#top 10 countries by world population percentage
df.sort_values(by="World Population Percentage", ascending=False).head(10)
```

```python
#analyzing correlations between populations
numeric_df = df.select_dtypes(include='number')
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```

```python
#visualizing correlations with a heatmap
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.rcParams['figure.figsize'] = (20,8)
plt.show()
```

```python
#analyzing average population by continent
df2 = pd.DataFrame(df)
result = (
    df.groupby('Continent')
    .mean(numeric_only=True)  
    .sort_values(by="2022 Population", ascending=False)
)
print(result)
```

```python
#average population trends by continent (1970–2022)
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)
df2
```

```python
#transposing the average population data
df3 = df2.transpose()
df3
```

```python
#visualizing population trends over time by continent
df3.plot(figsize=(12, 4))
```

```python
#boxplot to visualize data distribution and outliers
df.boxplot(figsize=(20,7))
plt.title("Boxplot of DataFrame Columns")
plt.show()
```
